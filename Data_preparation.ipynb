### Data preparation

```python
# read raw data
import pandas as pd

RawData_df = pd.read_json("../input/dm2022-isa5810-lab2-homework/tweets_DM.json", lines=True)
RawData_df
```

```python
# extract useful data
df = pd.json_normalize(RawData_df['_source'])
df.rename({'tweet.hashtags': 'hashtags', 'tweet.tweet_id': 'tweet_id', 'tweet.text': 'text'}, axis=1, inplace=True)
df
```

```python
# read identification data and merge with main data
data_identification = pd.read_csv('../input/dm2022-isa5810-lab2-homework/data_identification.csv') 
df = pd.merge(df, data_identification, on='tweet_id')
df
```

```python
df[df['identification'] == 'train']
```

```python
# read emotion data and merge with train data
df_train = df[df['identification'] == 'train']
emotion = pd.read_csv('../input/dm2022-isa5810-lab2-homework/emotion.csv') 
df_train = pd.merge(df_train, emotion, on='tweet_id')
df_train.drop('identification', axis=1, inplace=True)
df_train.reset_index(drop=True, inplace=True)
df_train
```

```python
# extract test data
df_test = df[df['identification'] == 'test']
df_test.drop('identification', axis = 1, inplace=True)
df_test.reset_index(drop=True, inplace=True)
df_test
```

#### save data to pickle

```python
df_test.to_pickle('df_test.pkl')
df_train.to_pickle('df_train.pkl')
```
